```python
type_of_vehicle_map = {"bicycle":1,"electric_scooter":2, "scooter": 3, "motorcycle" :4}
Weather_conditions_map = {'Sunny':1, 'Cloudy':2, 'Windy':3, 'Fog':4,'Sandstorms':5, 'Stormy':6 }
type_of_order_map = {'Snack':1 , "Meal": 2, 'Drinks': 3,'Buffet':4 }
Road_traffic_density_map = {'Low' :1, 'Medium' :2,"Jam":3, 'High' :4}
Festival_map = {'No':1, 'Yes':2}
City_map = {'Semi-Urban':1,"Urban":2 , "Metropolitian":3 }

```

```python
df['Type_of_vehicle'] = df["Type_of_vehicle"].replace(type_of_vehicle_map)
df['Weather_conditions'] = df["Weather_conditions"].replace(Weather_conditions_map)
df['Type_of_order'] = df["Type_of_order"].replace(type_of_order_map)
df['Road_traffic_density'] = df["Road_traffic_density"].replace(Road_traffic_density_map)
df['Festival'] = df["Festival"].replace(Festival_map)
df['City'] = df["City"].replace(City_map)
```
